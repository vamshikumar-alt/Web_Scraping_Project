```python
for i in range(0,29):
    print('https://www.reliancedigital.in/smart-phones/c/S101711?searchQuery=:relevance&page={}'.format(i))
```

```python
prices = []
details= []


for i in range(0,29):
    url = 'https://www.reliancedigital.in/smart-phones/c/S101711?searchQuery=:relevance&page={}'.format(i)
    
    page = requests.get(url)
    htmlCode = page.text
    
    soup = BeautifulSoup(htmlCode)
    
    for x in soup.find_all('div', attrs={'class' : 'sp grid'}):
        
        price = x.find('span',attrs={'class':'sc-bdVaJa hKEXmy'})
        if price is None:
            prices.append(np.NaN)
        else:
            prices.append(price.text)

        features = x.find('p',attrs={'class':'sp__name'})
        if features is None:
            details.append(np.NaN)
        else:
            details.append(features.text)
```

```python
print(len(prices))
print(len(details))
```

```python
df = pd.DataFrame({'prices' : prices, 'features' : details})
```
